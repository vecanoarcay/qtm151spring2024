```python
students = pd.read_csv("data_raw/students.csv")

bins_x = [0,54,59,64,69,74,79,82,86,92,100]

labels_x = ["F",
            "D",
            "C-",
            "C",
            "C+",
            "B-",
            "B",
            "B+",
            "A-",
            "A"]

students["lettergrade"] = pd.cut(students["numericgrade"],
                              bins = bins_x,
                              right = True,
                              labels = labels_x)

students

#ASK why is it giving me NA for 1
```
